```python
def split_coco_data(data, ids, split_name):
    images = []
    for img in data['images']:
        if img['id'] in ids:
            img_copy = img.copy()  # 复制避免修改原数据
            img_copy['file_name'] = f'{split_name}/{os.path.basename(img_copy["file_name"])}'
            images.append(img_copy)

    annotations = [ann for ann in data['annotations'] if ann['image_id'] in ids]

    split_data = {
        'info': data['info'],
        'licenses': data['licenses'],
        'images': images,
        'annotations': annotations,
        'categories': data['categories']
    }

    return split_data
```

```python
import os
import json

# 设置输出文件夹路径
output_folder = 'D:/comp9517/ass/project/archive (1)/turtles-data/data/annotations'

# 创建 output_folder 文件夹
os.makedirs(output_folder, exist_ok=True)

# 创建 train, valid, test 数据
train_data = split_coco_data(data, train_ids, 'train')
valid_data = split_coco_data(data, valid_ids, 'valid')
test_data = split_coco_data(data, test_ids, 'test')

# 保存为三个 JSON 文件
with open(os.path.join(output_folder, 'train.json'), 'w') as f:
    json.dump(train_data, f, indent=4)

with open(os.path.join(output_folder, 'valid.json'), 'w') as f:
    json.dump(valid_data, f, indent=4)

with open(os.path.join(output_folder, 'test.json'), 'w') as f:
    json.dump(test_data, f, indent=4)

print("数据集已成功分割并保存为 train.json, valid.json 和 test.json")
```

```python
import os
from PIL import Image
import torch
from torch.utils.data import Dataset
import torchvision.transforms as transforms

from torch.utils.data import Dataset
import os

class CustomCocoDataset(Dataset):
    def __init__(self, image_dir, ann_file, transform=None, target_transform=None):
        self.coco = COCO(ann_file)
        self.image_dir = image_dir
        self.transform = transform
        self.target_transform = target_transform
        self.ids = list(self.coco.imgs.keys())

    def __len__(self):
        return len(self.ids)

    def __getitem__(self, index):
        img_id = self.ids[index]
        ann_ids = self.coco.getAnnIds(imgIds=img_id)
        anns = self.coco.loadAnns(ann_ids)
        img_info = self.coco.loadImgs(img_id)[0]
        img_path = os.path.join(self.image_dir, img_info['file_name'])

        # 检查图像是否存在
        try:
            image = Image.open(img_path).convert("RGB")
        except FileNotFoundError:
            return None  # 返回 None 以跳过该图像

        # 创建空掩码并填充
        mask = np.zeros((img_info['height'], img_info['width']), dtype=np.uint8)
        for ann in anns:
            mask = np.maximum(mask, self.coco.annToMask(ann) * ann['category_id'])

        mask = Image.fromarray(mask)

        if self.transform:
            image = self.transform(image)
        if self.target_transform:
            mask = self.target_transform(mask)

        return image, mask
```

```python
image_transform = transforms.Compose([
    transforms.Resize((512, 512)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

mask_transform = transforms.Compose([
    transforms.Resize((512, 512), interpolation=Image.NEAREST),
    transforms.ToTensor(),
    transforms.Lambda(lambda x: torch.round(x * 255).long())
])
```

```python
from torch.utils.data import DataLoader
from torch.utils.data import DataLoader, Subset


image_dir = 'D:/comp9517/ass/project/archive (1)/turtles-data/data'
ann_file = 'D:/comp9517/ass/project/archive (1)/turtles-data//data/annotations.json'  # 指向 annotations.json
train_json = 'D:/comp9517/ass/project/archive (1)/turtles-data//data/annotations/train.json'
test_json = 'D:/comp9517/ass/project/archive (1)/turtles-data//data/annotations/test.json'
valid_json = 'D:/comp9517/ass/project/archive (1)/turtles-data//data/annotations/valid.json'


# 创建数据集
train_dataset = CustomCocoDataset(
    image_dir=image_dir,
    ann_file=train_json,
    transform=image_transform,
    target_transform=mask_transform
)

valid_dataset = CustomCocoDataset(
    image_dir=image_dir,
    ann_file=valid_json,
    transform=image_transform,
    target_transform=mask_transform
)

test_dataset = CustomCocoDataset(
    image_dir=image_dir,
    ann_file=test_json,
    transform=image_transform,
    target_transform=mask_transform
)

# subset_size = 10
# subset_indices = list(range(subset_size))
# subset_dataset = Subset(train_dataset, subset_indices)

# 创建数据加载器
train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True, num_workers=0)
valid_loader = DataLoader(valid_dataset, batch_size=4, shuffle=False, num_workers=0)
test_loader = DataLoader(test_dataset, batch_size=4, shuffle=False, num_workers=0)


```

```python
import torchvision.models.segmentation
import torch.optim as optim

num_classes = 4  # 假设包括背景在内有15个类别
model = torchvision.models.segmentation.deeplabv3_resnet50(pretrained=True)
model.classifier[4] = torch.nn.Conv2d(256, num_classes, kernel_size=(1, 1))  # 修改分类器的输出通道数

device = torch.device('cuda')
model = model.to(device)

criterion = torch.nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
import torch
from torch.utils.data import DataLoader

num_epochs = 10

print(f"Dataset size: {len(train_dataset)}")
print(f"Loader batch count: {len(train_loader)}")

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0

    for images, masks in train_loader:
        # 将图像和掩码加载到设备上
        images = images.to(device)
        masks = masks.to(device)

        # 检查图像和掩码尺寸是否符合要求
        if images.shape[-2:] != (512, 512) or masks.shape[-2:] != (512, 512):
            print("Skipping batch due to incorrect image or mask dimensions.")
            continue  # 跳过此批次

        # 检查掩码的唯一值，确保其符合类别要求
        mask_unique_values = masks.unique()
        if not all((0 <= val <= 255 for val in mask_unique_values)):
            print("Skipping batch due to unexpected mask values.")
            continue  # 跳过此批次

        # 将掩码转换为所需的整数格式
        masks = masks.squeeze(1).long()

        # 执行前向传播并获取输出
        optimizer.zero_grad()
        try:
            outputs = model(images)['out']
        except ValueError as e:
            print(f"Skipping batch due to model output shape error: {e}")
            continue  # 跳过此批次

        # 确保输出尺寸大于 1x1
        if outputs.shape[-2:] == (1, 1):
            print("Skipping batch due to output shape being too small.")
            continue  # 跳过此批次

        # 计算损失并进行反向传播和优化
        loss = criterion(outputs, masks)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss/len(train_loader)}")

print("Finished Training")
```

```python
import torch
import numpy as np

# IoU 计算函数
def compute_iou(pred, target, num_classes):
    ious = []
    for cls in range(num_classes):
        pred_inds = (pred == cls)
        target_inds = (target == cls)

        intersection = (pred_inds & target_inds).sum().float()
        union = (pred_inds | target_inds).sum().float()

        if union == 0:
            ious.append(float('nan'))  # 忽略没有出现的类
        else:
            ious.append((intersection / union).item())
    return ious

# 在验证集上评估 mIoU
def evaluate(model, dataloader, device, num_classes):
    model.eval()
    ious_all = []
    with torch.no_grad():
        for images, masks in dataloader:
            images = images.to(device)
            masks = masks.to(device)
            masks = masks.squeeze(1)  # 确保掩码尺寸为 (batch_size, H, W)

            # 获取模型预测
            outputs = model(images)['out']
            preds = torch.argmax(outputs, dim=1)  # 获取每个像素的预测类别

            # 计算每张图片的 IoU
            for pred, mask in zip(preds, masks):
                ious = compute_iou(pred, mask, num_classes)
                ious_all.append(ious)

    # 计算 mIoU，忽略没有出现的类
    ious_all = np.array(ious_all)
    mious = np.nanmean(ious_all, axis=0)  # 对每类求平均，忽略 NaN
    miou = np.nanmean(mious)  # 求所有类的均值 mIoU

    return miou, mious
```

```python
# 假设验证或测试集的 DataLoader 为 val_loader
num_classes = 4  # 根据您的数据集中的类别数调整
miou, ious_per_class = evaluate(model, test_loader, device, num_classes)

print(f"Mean IoU (mIoU): {miou}")
for i, iou in enumerate(ious_per_class):
    print(f"IoU for class {i}: {iou}")
```
